```python
# importing the required libraries 
import pandas as pd 
import requests
import mysql.connector 
```

```python
mydb = mysql.connector.connect(host='localhost', port='3306',user='root',password='****',database='world')
```

```python
mydb
```

```python
pd.read_sql_query('SHOW TABLES', mydb)  # call to check all the tables in the database
```

```python
pd.read_sql_query('SELECT * FROM country LIMIT 10 ',mydb).head()
```

```python
df = pd.read_sql_query('SELECT * FROM country', mydb)
```

```python
df.head()
```

```python
df.query('GovernmentForm == "Republic"').head()
```

```python
df.query('HeadOfState =="Elisabeth II"').head()
```

```python
pd.read_sql_query('SELECT * FROM country WHERE HeadOfState = "Elisabeth II"',mydb).head() # producing same output using SQL
```

```python
# countries which have government forms as republics 

pd.read_sql_query('SELECT * FROM country WHERE GovernmentForm REGEXP "republic"', mydb).head()
```
